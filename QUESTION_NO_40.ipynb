# Find the Median of a List

```python
def find_median(nums_list: list) -> float:
    if not isinstance(nums_list, list):
        raise ValueError("Input must be a list")

    if len(nums_list) == 0:
        raise ValueError("List cannot be empty")

    for num in nums_list:
        if not isinstance(num, (int, float)):
            raise ValueError("All elements of the list must be int or float")

    nums_list.sort()
    list_length = len(nums_list)
    mid = list_length // 2

    if list_length % 2 != 0:
        return nums_list[mid]
    else:
        return (nums_list[mid - 1] + nums_list[mid]) / 2


if __name__ == "__main__":
    try:
        print(find_median([7, 2, 3, 1, 5, 4, 6]))
    except (TypeError, ValueError) as e:
        print(e)
    except Exception as e:
        print(e)
```
