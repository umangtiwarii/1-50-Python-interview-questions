# Convert List of Tuples to Dictionary

```python
def list_to_dict(List : list) -> dict:
    my_dict = {}
    for key , value in List:
        my_dict [key] = value
    return my_dict
if __name__ == "__main__":
    data = [('a' , 1) , ('b' , 2) , ('c' , 3) , ('d' , 6)]
    try:
        print(list_to_dict(data))
    except Exception as e:
        print (e)
```
