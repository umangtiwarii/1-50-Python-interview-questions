# Find Sum of List Using Recursion

```python
def recursive_sum(list1: list) -> int:
    if len(list1) == 0:
        return 0
    return list1[0] + recursive_sum(list1[1:])
    if not isinstance(list1,list):
        raise TypeError ("Input must be a list")
    if len(list1) == 0:
        raise ValueError ("list can't be empty.")
    if not isinstance(list1[0],(int,float)):
        raise TypeError("List can contain only numeric values.")
    if len(list1) == 1:
        return list1[0]

    if len(list1) == 0:
        return 0

    return list1[0] + recursive_sum(list1[1:])
if __name__ == "__main__":
    list1 = [12,177]
    try:
        print(recursive_sum(list1))
    except (TypeError,ValueError) as e:
        print(e)
```
