# Find Pairs with Given Sum in a List 

```python
def find_pairs(List : list , target : int) -> list:
    if not isinstance(target,(int,float)):
        raise TypeError("Target must be int or float.")
    if not isinstance(List,list):
        raise TypeError("First argument must be a list.")
    for x in List:
        if not isinstance(x,(int,float)):
            raise TypeError (f"{x} is not a valid input , all elements must be int or float")
    pairs = []
    for i in range(len(List)):
        for j in range(i + 1, len(List)):
            if List[i] + List[j] == target:
                pairs.append((List[i],List[j]))
    return pairs

if  __name__ == "__main__":
    try:
        
        List = [1,2,3,1,4,5,6,7,8,9,13]
        target = 22
        print(find_pairs(List,target))
    except TypeError as e:
        print(e)
    except Exception as e:
        print(e)
```
