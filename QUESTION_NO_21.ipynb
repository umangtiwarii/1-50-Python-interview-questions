# Python Program to Find GCD Using Recursion

```python
def recursive_gcd(a: int, b: int) -> int:

    if not isinstance(a, int) or not isinstance(b, int):
        raise TypeError("Both inputs must be integers.")

    if a == 0 and b == 0:
        raise ValueError("Both numbers cannot be zero.")

    a, b = abs(a), abs(b)

    if b == 0:
        return a

    return recursive_gcd(b, a % b)


if __name__ == "__main__":

    first_number = 200
    second_number = 2932

    try:

        result = recursive_gcd(first_number, second_number)

        print("GCD is:", result)

    except (TypeError, ValueError) as e:
        print(e)

    except Exception as e:
        print(e)
```
