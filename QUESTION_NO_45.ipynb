# Find Common Elements in 3 Lists

```python
def commom_finder(list1 : list , list2 : list , list3 : list) -> list:
    if not isinstance (list1 , list) or not isinstance (list2 , list) or not isinstance (list3 , list):
        raise TypeError ("Input must be a list")
    common = []
    for value in list1:
        if value in list2 and value in list3 and value not in common:
            common.append(value)
    return common

if __name__ == "__main__":
    try:
        list1 = [1,2,3,4,5,6,7,8]
        list2 = [1,2,3,4,5,6,7,8]
        list3 = [1,2,3,4,5,6,7]
        print(commom_finder(list1 , list2 , list3))
    except TypeError as e:
        print(e)
    except Exception as e:
        print(e)
```
