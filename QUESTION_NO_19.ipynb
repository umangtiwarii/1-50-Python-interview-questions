# Reverse Words in a Sentence 

```python
def rev_sentence(sentence : str)-> str:
    if not isinstance(sentence , str):
        raise TypeError("input muts be a string")
    word_list = []
    word = ""
    for char in sentence:
        if char != " ":
            word = word + char
        else:
            word_list.append(word)
            word = ""
    word_list.append(word)
    rev = ""
    for i in range(len(word_list)-1,-1,-1):
        rev = rev + word_list[i]
        if i != 0:
            rev = rev + " "
    return rev
if __name__ == "__main__":
    my_sentence = "this is python"
    try:
        print(rev_sentence(my_sentence))
    except TypeError as e:
        print(e)
    except Exception as e:
        print(e)
```

```python
text = "python is fun"

result = ""
word = ""

for ch in text + " ":   # extra space to capture last word
    if ch != " ":
        word += ch
    else:
        result = word + (" " + result if result else "")
        word = ""

print(result)
```
