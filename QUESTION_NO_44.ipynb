# How to Swap Two Variables in Python Without Temporary Variable

```python
def value_swap (a,b):
    a,b = b,a
    return a,b

if __name__ == "__main__":
    try:
        a = 5
        b = 10
        print("value before swap: " , a)
        print("value before swap: " , b)
        a,b = value_swap (a,b)
        print("value after swap : " , a)
        print("Value after swap : " , b)
    except Exception as e:
        print(e)
```
