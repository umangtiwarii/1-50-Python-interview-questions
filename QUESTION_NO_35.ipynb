# Check Power of 2 in Python Using Bitwise Operations

```python
def power_of_two_info(n):
    if not isinstance(n, int) or n <= 0:
        return False, -1

    power = 0
    while n > 1 and n % 2 == 0:
        n //= 2
        power += 1

    return (n == 1, power if n == 1 else -1)


if __name__ == "__main__":
    num = int(input("Enter a number: "))
    print(power_of_two_info(num))
```
