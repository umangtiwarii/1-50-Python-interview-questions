# Check if a String Contains Only Digits

```python
def check_digits (digit : str)-> bool:
    for num in digit:
        if num < "0" or num > "9":
            return False
    return True
if __name__ == "__main__":
    try:
        digit = "62378"
        print(check_digits(digit))
    except Exception as e:
        print(e)
```
