# Reverse an Array Without Extra Space

```python
def reverse_array(nums : list) ->list:
    if not isinstance(nums , list):
        raise TypeError("Input must be a list")
    if len(nums) < 2:
        raise ValueError ("Atleast 2 values are required")
    left = 0
    right = len(nums)-1
    while left < right:
        nums[left] ,nums[right] = nums[right] , nums[left]
        left = left +1
        right = right -1
    return nums
if __name__ == "__main__":
    nums = ["Pyth","pytho"]
    try:
        print(reverse_array(nums))
    except (TypeError,ValueError) as r:
        print(r)
    except Exception as e:
        print(e)
```
