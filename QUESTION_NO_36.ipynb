# How to Flatten a Nested List

```python
def flatten_list(nested : list) -> list:
    result = []
    temp_list = list(nested)
    while temp_list:
        item = temp_list.pop()
        if isinstance(item , list):
            temp_list.extend(item)
        else:
            result.append(item)
    #     print("result : " , result)
    # print("final_result " , result)
    return result[::-1]
if __name__ == "__main__":
    try:
        print(flatten_list([1,2,[3,[4,5,6],7,],8]))
    except Exception as e:
        print(e)
```
