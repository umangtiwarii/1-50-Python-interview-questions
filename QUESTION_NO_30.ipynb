# Find missing number in a list

```python
def find_missing(list1 : list) -> int:
    if not isinstance (list1 , list):
        raise TypeError ("Input must be a list")
    for item in list1:
        if not isinstance(item , int):
            raise ValueError("all items must be integers only")
    for item in range(0,len(list1)-1):
        if list1[item+1] - list1[item] != 1:
            return list1[item]+1

if __name__ == "__main__":
    list1 = [1,2,3,4,6,7,8,9]
    try:
        print(find_missing(list1))
    except (TypeError,ValueError) as e:
        print(e)
    except Exception as e:
        print(e)
```

```python
def missing_value(sample_list:list)->int:
    if not isinstance(sample_list,list):
        raise TypeError("Input must be a list")
    if len(sample_list) < 2:
        raise ValueError("Atleast two numbers required inside the list")
    if not all(isinstance(item,int) for item in sample_list):
        raise TypeError("All elements of list must be integers")
    sample_list = sorted(sample_list)
    for i in range(0,len(sample_list)-1):
        if sample_list[i+1] - sample_list[i] != 1:
            return sample_list[i]+1
            break
    else:
        return None
if __name__ == "__main__":
    sample_list = [12,14,15,16,17,18,19,20]
    try:
        print(missing_value(sample_list))
    except(TypeError,ValueError) as e:
        print(e)
    except Exception as e:
        print("something went wrong",e)
```
