# Remove Nth occurrence of the character from a string

```python
def remove_nth_occurrence(string : str , char_to_remove : str , n : int)->str:
    if not isinstance(string,str) or not isinstance(char_to_remove , str) or not isinstance(n , int):
        raise TypeError ("2 arguments must be string and 3rd argument must be an integer")
    occurrence_count = 0
    result_characters = []
    for char in string:
        if char == char_to_remove:
            occurrence_count = occurrence_count + 1
            if occurrence_count == n:
                continue
                
        result_characters.append(char)
    if occurrence_count < n:
        raise ValueError(f"{char_to_remove} does not occur {n} times in the string")
    return "".join(result_characters)
    
if __name__ == "__main__":
    
    try:
        
        user_input1 = input("enter your text")
        user_input2 = input("character to remove")
        user_input3 = int(input("which occurrence should be removed"))
        result = remove_nth_occurrence(user_input1 , user_input2 , user_input3)
        print(result)
        
    except TypeError as e:
        print(e)
    except Exception as e:
        print(e)
        
```
