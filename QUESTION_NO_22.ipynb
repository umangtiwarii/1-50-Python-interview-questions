# Convert Decimal to Binary Without Using bin()

```python
def decimal_to_binary(decimal : int) -> str:
    if not isinstance(decimal , int):
        raise TypeError("Input must be a integer value")
    if decimal == 0:
        return "0"

    binary = ""
    while decimal > 0:
        remainder = decimal % 2
        binary = str(remainder) + binary
        decimal = decimal // 2
    return binary
if __name__ == "__main__":
    decimal = 1234567
    try:
        print(decimal_to_binary(decimal))
    except TypeError as e:
        print(e)
    except Exception as e:
        print(e)
```
