# Armstrong Number

```python
def is_armstrong(value : int) -> bool:
    temp_copy = value
    digit_count = 0
    while temp_copy > 0:
        digit_count += 1
        temp_copy = temp_copy //10
    total = 0
    temp_copy = value
    while temp_copy > 0:
        digit = temp_copy % 10
        total = total + digit ** digit_count
        temp_copy = temp_copy //10
    return total == value
```

```python
def is_armstrong(value: int) -> bool:
    temp_copy = value
    digit_count = 0

    while temp_copy > 0:
        digit_count += 1
        temp_copy = temp_copy // 10

    total = 0
    temp_copy = value

    while temp_copy > 0:
        digit = temp_copy % 10
        total = total + digit ** digit_count
        temp_copy = temp_copy // 10

    return total == value


num = int(input())

if is_armstrong(num):
    print("It's an Armstrong number")
else:
    print("Not an Armstrong number")
```
