# FIND FACTORIAL OF A NUMBER

```python
def My_factorial(num : int) -> int:
    if isinstance(num,bool):
        raise TypeError("boolean values are not allowed")
    if not isinstance(num,int):
        raise TypeError("input must be an integer")
    if num < 0:
        raise ValueError("factorial is not defined for negetive numbers")
    factorial = 1
    for i in range(1,num+1):
        factorial = factorial * i
    return factorial
if __name__ == "__main__":
    try:
        number = input("enter your number to find factorial: ")
        number = int(number)
        print("factorial",My_factorial(number))
    except TypeError as e:
        print(e)
    except ValueError as e:
        print("input must be an int")
```
