# FIND VOWELS IN A STRING

```python
def my_vowels(string : str) -> str:
    if string.isdigit():
    # if not isinstance(string,str):
        raise TypeError("Input must be string.")
    vowels = "aeiou"
    count = 0
    for char in string.lower():
        if char in vowels:
            count = count + 1
    return count
if __name__ == "__main__":
    try:
        user_input = input("Enter your text to check the count of vowels in your text: ")
        result = my_vowels(user_input)
        print(f"There are `{result}` vowels in your text.")
    except TypeError as e:
        print(e)
    except Exception as e:
        print(e)
```
